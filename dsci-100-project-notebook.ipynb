#### Task 4.2: Determine the amount of missingness (2 points)

- Use your `tidyverse` knowledge to create **one** table that calculates the number of missing values for each column in the data set (both numeric and categorical).
- Show the outputted table. 
</div>

```r
missing_value <- woman_basketball_selected |>
  summarise(across(everything(), ~
                   sum(is.na(.)))) |>
  pivot_longer(everything(),
               names_to = "column",
               values_to = "num_missing")
missing_value
```

However, KNN has limitations. It can perform poorly with very small datasets, and my dataset contains only 20 observations. Because KNN relies on distances, it is also sensitive to outliers and requires predictors to be standardized. Additionally, KNN can overfit if k is too small or underfit if k is too large, so tuning is important.
